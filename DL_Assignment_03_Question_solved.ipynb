```python
import numpy as np
import pandas as pd

df = pd.read_csv("data.csv")
df.head()
```

```python
df = df.drop(columns=["id", "Unnamed: 32"])
print(df.isnull().sum())
```

```python
df["diagnosis"] = df["diagnosis"].map({"M": 0, "B": 1})
X = df.drop(columns=["diagnosis"])
y = df["diagnosis"]

X = X.values
y = y.values.reshape(-1, 1)

X = (X-X.mean(axis=0))/X.std(axis=0)

split = int(0.8*len(X))

X_train = X[:split]
y_train = y[:split]

X_test = X[split:]
y_test = y[split:]

```

```python
import torch

X_train = torch.tensor(X_train,dtype=torch.float32)
y_train = torch.tensor(y_train,dtype=torch.float32)

X_test = torch.tensor(X_test,dtype=torch.float32)
y_test = torch.tensor(y_test,dtype=torch.float32)
```

Removed columns (id, Unnamed: 32) → not useful for prediction

Checked missing values → none found, so no action needed

Encoded target (M, B) → (0, 1) → models need numeric data

Separated features (X) and target (y) → required for supervised learning

Converted to NumPy arrays → efficient computation

Applied feature scaling → ensures all features contribute equally

Split data (80% train, 20% test) → evaluate model on unseen data

Converted to PyTorch tensors → required for model training

## Design a neural network using nn.Module.

## Write Answer 02:


```python
import torch.nn as nn

class BreastCancerNN(nn.Module):
    def __init__(self, input_size):
        super(BreastCancerNN, self).__init__()

        self.fc1 = nn.Linear(input_size, 16)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(16, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.sigmoid(x)
        return x
```

```python
model = BreastCancerNN(input_size=X_train.shape[1])
print(model)
```

```python
total_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print("Total Trainable Parameters:", total_params)
```

2. Activation Function

# Question 04: [ Marks 15 ]

## Implement a full training loop:

Forward pass

Loss computation

Backward pass

Parameter update

Gradient reset

### Requirements:

Train for at least 100 epochs.

Print loss every 10 epochs.

Store training loss history(You can pick your own Data Structure).

Explain clearly what happens in each step of the pipeline.

## Write Answer 04:

```python
num_epochs = 100
loss_history = []

for epoch in range(num_epochs):

    outputs = model(X_train)

    loss = criterion(outputs, y_train)

    loss.backward()

    optimizer.step()

    optimizer.zero_grad()

    loss_history.append(loss.item())

    if (epoch + 1) % 10 == 0:
        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")
```

Forward pass: Model takes input and produces predictions

Loss computation: Calculates error between prediction and actual value

Backward pass: Computes gradients (how to fix the error)

Parameter update: Optimizer updates weights to reduce loss

Gradient reset: Clears old gradients to avoid accumulation

Loss history: Stores loss to track learning progress

```python
#modified(Hidden neurons: 32)

class ModifiedNN(nn.Module):
    def __init__(self, input_size):
        super(ModifiedNN, self).__init__()

        self.fc1 = nn.Linear(input_size, 32)  # changed from 16 → 32
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(32, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.sigmoid(x)
        return x
```

```python
model = ModifiedNN(input_size=X_train.shape[1])

criterion = nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
num_epochs = 100
loss_history_new = []

for epoch in range(num_epochs):

    outputs = model(X_train)
    loss = criterion(outputs, y_train)

    loss.backward()
    optimizer.step()
    optimizer.zero_grad()

    loss_history_new.append(loss.item())

    if (epoch + 1) % 10 == 0:
        print(f"Epoch [{epoch+1}/100], Loss: {loss.item():.4f}")
```

```python
with torch.no_grad():
    y_pred = model(X_test)
    y_pred_label = (y_pred > 0.5).float()

    test_acc_new = (y_pred_label == y_test).sum().item() / len(y_test)

print("New Test Accuracy:", test_acc_new)
```

Increasing the number of hidden neurons from 16 to 32 improved the model’s capacity to learn complex patterns, resulting in better feature representation and higher test accuracy. This indicates that the previous model was slightly underfitting, and the increased neurons helped achieve better generalization without causing overfitting.

# Training Analysis

Answer the following:

Why must gradients be reset every epoch?

What happens if learning rate is too high?

What happens if learning rate is too small?

Why do we define layers inside the constructor (__init__) and not inside forward()?


## Write Answer 07:

Why must gradients be reset every epoch?

Gradients accumulate. Resetting them prevents previous epoch's gradients from influencing current updates, ensuring correct learning.

What happens if the learning rate is too high?

Too high a learning rate causes the model to overshoot the minimum, bounce erratically, or diverge, leading to unstable training and increased loss.

What happens if the learning rate is too small?

Too small a learning rate makes training very slow, potentially trapping the model in local minima or plateaus, and delaying convergence significantly.

Why do we define layers inside the constructor (__init __) and not inside forward()?

Layers are model components with learnable parameters, defined once in __init __ for initialization. forward() defines the data flow through these already-initialized layers; defining layers there would re-create them repeatedly, preventing learning and wasting memory.
